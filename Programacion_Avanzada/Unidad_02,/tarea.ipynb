### 2. Crear una clase **`Funcion`**

```python
print(funciones.Funcion.__doc__)
```

```python
# Creo funciones para generar las visualizaciones en el punto 3.
f = funciones.Funcion("f", ["x"], lambda x: x ** 2)
g = funciones.Funcion("g", ["x", "y"], lambda x, y: np.cos(x) * np.sin(x))
```

### 3. Visualizaciones

```python
# Evaluar funci√≥n f(x) = x^2 creada en el punto 2
x_vals = np.linspace(-10, 10, 100)
f_vals = f.evaluar(x_vals)

plt.figure(figsize = (6, 5))
plt.plot(x_vals, f_vals, label=r"$f(x) = x^2$", color = 'b')
plt.xlabel(r"$x$")
plt.ylabel(r"$f(x)$")
plt.title(r"$f(x) = x^2$")
plt.legend()
plt.grid()
plt.savefig('./graficos/grafico_2d.png', dpi = 300)
plt.show()
```

```python
# Evaluar funci√≥n g(x, y) = cos(x) * sin(y) en el punto 2
x_vals = np.linspace(-5, 5, 50)
y_vals = np.linspace(-5, 5, 50)
Z = g.evaluar(x_vals, y_vals)

X, Y = np.meshgrid(x_vals, y_vals, indexing = 'ij')

fig = plt.figure(figsize = (10, 6))
ax = fig.add_subplot(111, projection = '3d')
ax.plot_surface(X, Y, Z, cmap = 'coolwarm')

ax.set_xlabel(r"$x$")
ax.set_ylabel(r"$y$")
ax.set_zlabel(r"$g(x, y)$")
ax.set_title(r"$g(x, y) = cos(x) sin(y)$")

plt.savefig('./graficos/grafico_3d.png', dpi = 300)
plt.show()
```
