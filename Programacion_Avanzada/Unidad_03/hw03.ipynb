* Debe poder sumar y restar puntos, ademas multiplicar por un escalar, es decir 
```python

C.x, C.y
```

````python
# Solucion 
A = Punto2D(3, 4)
B = Punto2D(5, 6)

# Multiplicacion por un escalar 
C = 3*A 
print(C)

# Suma de dos puntos 
D = C + B 
print(D)

# Resta de dos Puntos 
E = C - B
print(E)
````

* Crear un método que devuelva la distancia entre dos puntos.
```python
A = Punto2D(3,4)
B = Punto2D(5,6)
A.distancia(B) 
```
Use el decorador para calcular el tiempo que se demora en calcular la distancia.

````python
# Solucion
A = Punto2D(3, 4)
B = Punto2D(5, 6)

# Calcular la distancia entre A y B
distancia = A.distancia(B)
print(f"La distancia entre {A} y {B} es: {distancia:.2f}")
````

* Que retorne la distancia al origen usando `abs`:
```python

``` 

````python
# Solucion
A = Punto2D(3, 4)
distancia_origen = abs(A)
print(f"La distancia de {A} al origen es: {distancia_origen:.2f}")
````
